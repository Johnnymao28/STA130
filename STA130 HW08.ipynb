## 1(b)

### A classification decision tree classifies data step by step using a tree-like structure. The decision tree begins at the root node, where it splits the data into two or more groups based on the values of a feature. It then moves down the branches, examining another feature at each child node to further divide the data. Ultimately, the data will reach a leaf node, which contains a category label. Therefore, the prediction process of a decision tree involves traversing the input data down the branches of the tree until it reaches a leaf node, where the corresponding category is output.

### Key difference between a Classification decision tree and Multipule linear regression:

### A classification decision tree is based on a series of conditional decisions to progressively group the data, ultimately predicting the outcome through majority voting among categories. It then returns a class label as the prediction. In contrast, multiple linear regression assumes that the relationships between variables are linear and uses a linear equation to compute a continuous output value. The key differences between the two lie in their decision-making processes and output types.

# Question 2

## Accuracy

### Accuracy represents the proportion of correct predictions among all predictions, including true positives (TP) and true negatives (TN). It is suitable for evaluating the overall performance of a model when the positive and negative sample proportions are balanced.

### Accuracy = (TP+TN)/(TP+TN+FP+FN)

### The application of accuracy in weather forecasting: predicting whether a day will be sunny or cloudy. Suppose there are 100 days of predictions, with 90 correct predictions (including both sunny and cloudy days) and 10 incorrect ones, resulting in an accuracy of 90%. Since the positive and negative classes (sunny and cloudy) are generally balanced, accuracy effectively reflects the overall performance of the predictions.

### The proportion of correct predictions among all predictions.

## Sensitivity

### Sensitivity represents the proportion of actual positive cases that are correctly predicted as positive. It focuses on the rate of missed cases and is suitable for scenarios where minimizing false negatives is critical, such as disease detection.

### Sensitivity = TP/(TP+FN)

### The application of sensitivity in cancer screening: detecting whether a patient has cancer. Suppose there are 100 cancer patients, and the model correctly identifies 90 of them while missing 10, resulting in a sensitivity of 90%. Missing cancer patients (false negatives) could lead to severe consequences, making sensitivity critically important in medical diagnosis.

### The ability to identify positive cases without missing any.

## Specificity

### Specificity represents the proportion of actual negative samples that are correctly predicted as negative. It focuses on the false positive rate and is suitable for scenarios where reducing false positives is important, such as dangerous goods detection.

### Specificity = TN/(TN+FP)

### The application of specificity in airport security screening for dangerous items: detecting whether luggage contains dangerous items. Suppose there are 100 safe pieces of luggage, and the model correctly classifies 90 of them while falsely identifying 10 as containing dangerous items, resulting in a specificity of 90%. Reducing false positives helps avoid unnecessary rechecks of normal passengers' luggage, saving resources and improving efficiency.

### The ability to exclude negative cases without making false positives.

## Precision

### Precision represents the proportion of predicted positive samples that are actually positive. It focuses on the false positive rate and is suitable for scenarios where ensuring the accuracy of positive predictions is crucial, such as spam email filtering.

### Precision = TP/(TP+FP)

### The application of precision in spam email filtering: predicting whether an email is spam. Suppose the system identifies 50 emails as spam, of which 40 are correctly classified, and 10 are false positives, resulting in a precision of 80%. Misclassifying important emails as spam (false positives) can cause inconvenience, making precision particularly important in email filtering.

### The proportion of actual positives among the predicted positives.

### In linear regression models, coefficients directly represent the impact of a specific feature on the predicted outcome, which is highly intuitive. The magnitude and sign of the coefficients indicate the direction and extent of the feature's influence on the predicted value.

### In decision trees, feature importance does not intuitively reflect a feature's direct impact on the predicted outcome, as it is calculated relative to other features. Feature importance represents a feature's relative contribution to the model's performance across all splits in the decision tree. Unlike linear regression coefficients, feature importance does not indicate a clear "direction" or the exact magnitude of the impact.
